```python
def print_unique_values_and_counts(df, column_name):
    """
    Print unique values and their counts for a specified column in a pandas DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.
    column_name (str): The name of the column to analyze.
    """
    if column_name in df.columns:
        value_counts = df[column_name].value_counts()
        print(f"Unique values and counts for column '{column_name}':\n{value_counts}")
    else:
        print(f"Column '{column_name}' not found in DataFrame.")
```

```python
def count_aggregate_values(df, column_name):
    """
    Return a count aggregate of all different values in a specified column of a pandas DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.
    column_name (str): The name of the column to analyze.

    Returns:
    pd.Series: A series containing counts of unique values in the specified column.
    """
    if column_name in df.columns:
        return len(df[column_name].value_counts())
    else:
        raise ValueError(f"Column '{column_name}' not found in DataFrame.")
```

```python
def aggregate_and_iterate(df, group_column, agg_column):
    """
    Aggregate values from a DataFrame on a given column and iterate over the results.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.
    group_column (str): The name of the column to group by.
    agg_column (str): The name of the column to aggregate.

    Returns:
    None
    """
    aggregated_df = df.groupby(group_column)[agg_column].sum().reset_index()
    for index, row in aggregated_df.iterrows():
        print(f"Group: {row[group_column]}, Aggregated Value: {row[agg_column]}")
```

```python
import pandas as pd
from pygbif import species


def fetch_gbif_results(names_list):
    """
    Fetch results from GBIF for each name in the list and create a pandas DataFrame.

    Parameters:
    names_list (list): A list of scientific names to search in GBIF.

    Returns:
    pd.DataFrame: A DataFrame containing the results from GBIF.
    """
    results = []

    for name in names_list:
        try:
            result = species.name_suggest(q=name)
            if result:
                for entry in result:
                    # Extract relevant fields from the entry
                    result_data = {
                        "sonotheque_scientific_name": name,
                        "key": entry.get("key"),
                        "nameKey": entry.get("nameKey"),
                        "vernacular_name": entry.get("vernacularNames"),
                        "kingdom": entry.get("kingdom"),
                        "phylum": entry.get("phylum"),
                        "class": entry.get("class"),
                        "order": entry.get("order"),
                        "family": entry.get("family"),
                        "genus": entry.get("genus"),
                        "species": entry.get("species"),
                        "scientific_name": entry.get("scientificName"),
                        "canonical_name": entry.get("canonicalName"),
                        "vernacular_name_original": entry.get(
                            "vernacularNamesOriginal"
                        ),
                        "vernacular_name_english": entry.get("vernacularNamesEnglish"),
                        "vernacular_name_french": entry.get("vernacularNamesFrench"),
                        "rank": entry.get("rank"),
                        "status": entry.get("status"),
                        "synonym": entry.get("synonym"),
                    }
                    results.append(result_data)
        except Exception as e:
            print(f"Error fetching data for {name}: {e}")

    # Convert the list of dictionaries to a DataFrame
    df_results = pd.DataFrame(results)
    return df_results
```

```python
scientific_names = df["NOM_SCIENTIFIQUE"].unique().tolist()
df_gbif_results = fetch_gbif_results(scientific_names)
print(df_gbif_results.head())
```

```python
import re
from rapidfuzz import fuzz, process


def custom_scorer(s1, s2):
    """
    Custom scoring function that considers both integer and non-integer parts of the strings.

    Parameters:
    s1 (str): The first string to compare.
    s2 (str): The second string to compare.

    Returns:
    int: The similarity score between the two strings.
    """
    # Extract integer parts
    int_part1 = "".join(re.findall(r"\d+", s1))
    int_part2 = "".join(re.findall(r"\d+", s2))

    # Extract non-integer parts
    non_int_part1 = re.sub(r"\d+", "", s1)
    non_int_part2 = re.sub(r"\d+", "", s2)

    # Calculate similarity scores
    int_score = fuzz.ratio(int_part1, int_part2)
    non_int_score = fuzz.ratio(non_int_part1, non_int_part2)

    # Combine scores with more weight to integer parts
    combined_score = 0.6 * int_score + 0.4 * non_int_score
    return combined_score


def deduplicate_list(strings_list, threshold=90):
    """
    Deduplicate very close entries in a list of strings using a custom scoring function.

    Parameters:
    strings_list (list): The list of strings to deduplicate.
    threshold (int): The similarity threshold for deduplication (default is 90).

    Returns:
    list: A deduplicated list of strings.
    """
    # Use rapidfuzz to deduplicate with the custom scorer
    deduped_list = list(
        process.dedupe(strings_list, scorer=custom_scorer, score_cutoff=threshold)
    )

    return deduped_list
```

```python
class DataExplorer:
    def __init__(self, df):
        self.df = df
        self.grouped_df = None
    
    def group_by(self, grouper_column_name, aggregated_column_names):
        # Create a dataframe that only contains columns listed in aggregated_column_names and grouper_column_name
        filtered_df = self.df[[grouper_column_name] + aggregated_column_names]
        grouped_df = (
            filtered_df.groupby(grouper_column_name)
            .agg(lambda x: x.tolist())
            .map(lambda x: list(set(x)))
            # .map(lambda x: x[0] if len(x) == 1 else x)
            .map(
                lambda x: pd.NA
                if isinstance(x, list) and all(pd.isna(item) for item in x)
                else x
            )
        )
        # Reset the index to make 'son_primary_key' a regular column again
        grouped_df = grouped_df.reset_index()
        print(grouped_df.head(10))
        self.grouped_df = grouped_df
        return self.grouped_df
    
    def get_unique_values(self, column_name):
        return self.df[column_name].unique().tolist()
```

```python
import polars as pl

path = "./data/sonotheque1.csv"
conversion_dict = {
    "SON_PK": "son_primary_key",
    "SON_UID": "code",
    "TYPE": "type",
    "TITRE": "title",
    "INSTITUTION_CODE": "institution_code",
    "COLLECTION_CODE": "collection_code",
    "CATALOG_NUMBER": "catalog_number",
    "VALIDE": "valid",
    "VALIDE_PAR": "validated_by",
    "PUBLIE": "published",
    "DATE_PUBLICATION_DEBUT": "publication_start_date",
    "DATE_PUBLICATION_FIN": "publication_end_date",
    "PUBLICATION_SCIENTIFIQUE": "scientific_publication",
    "RECORDER": "recorder",
    "MICRO": "micro",
    "PARABOLE": "parabola",
    "FREQUENCE_ECHANTILLONNAGE": "sampling_frequency",
    "SUPPORT_ORIGINAL": "original_support",
    "QUALITE": "quality",
    "DATE_ENREGISTREMENT_INT": "registration_date_int",
    "HEURE_ENREGISTREMENT_INT": "registration_time_int",
    "LIEU_ENREGISTREMENT": "registration_place",
    "TEMPERATURE": "temperature",
    "COMMENTAIRES_ENREGISTREMENT": "registration_comments",
    "CONTEXTE_COMPORTEMENTAL": "behavioral_context",
    "HABITAT": "habitat",
    "DESCRIPTION_HABITAT": "habitat_description",
    "CONTINENT_OCEAN": "continent_ocean",
    "PAYS": "country",
    "LOCALITE": "locality",
    "LATITUDE": "latitude",
    "LONGITUDE": "longitude",
    "ALTITUDE": "altitude",
    "COMMENTAIRES": "comments",
    "DATECRE": "creation_date",
    "DATEMAJ": "last_modified_date",
    "USERCRE": "created_by",
    "USERMAJ": "modified_by",
    "NOM_SCIENTIFIQUE": "scientific_name",
    "ESPECE_PK": "species_primary_key",
    "MIME_TYPE": "mime_type",
}

# Scan the CSV file with Polars
lf = pl.scan_csv(path, separator=";", infer_schema_length=None)

# Rename columns using the conversion dictionary and remove columns not listed in the dictionary
lf = lf.rename(conversion_dict).select([pl.col(new_name) for new_name in conversion_dict.values()])

# Convert "last_modified_date" and "creation_date" columns to Polars temporal Datetime format
lf = lf.with_columns([
    pl.col("last_modified_date").str.strptime(pl.Datetime, "%Y-%m-%d %H:%M:%S%.3f").alias("last_modified_date"),
    pl.col("creation_date").str.strptime(pl.Datetime, "%Y-%m-%d %H:%M:%S%.3f").alias("creation_date"),
    pl.col("mime_type").cast(pl.Categorical),
    pl.col("type").cast(pl.Categorical)
])


# Print the schema of the lazy frame
print(lf.head(10).collect())

```
